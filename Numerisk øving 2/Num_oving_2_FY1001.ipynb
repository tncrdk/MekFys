```python
import matplotlib.pyplot as plt
import numpy as np

def f(A,p,x):
    return A*np.sin(x+p)

A = 5
p = np.pi/3
a = 25
l = np.linspace(0,3*np.pi,a)
b = f(A,p,l)
%matplotlib inline
plt.figure(0)
plt.plot(l,b)
plt.title('Funksjonen')
plt.show()
```

```python
"""Code for calculating and illustrating the sine-wave (equation 1.8) in exercises 3 FY0110
   Per Persson 1/7/19
"""

import matplotlib.pyplot as plt
import numpy as np

def sine_wave(amplitude, phase, x_values):
    """ Calculate the sine wave and return the result (equation 1.8)
    """
    return amplitude*np.sin(x_values + phase)

amplitude = 5  
phase = np.pi/3
x_start = 0
x_end = 3*np.pi
num_x_values = 25  # Number of x-values from x_start to x_end

x_values = np.linspace(x_start, x_end, num_x_values)  # Create list of x-values
sine_res = sine_wave(amplitude, phase, x_values)  # Calculate sine wave for x-values

%matplotlib inline
plt.figure(0)
plt.plot(x_values, sine_res)  # Plot result
plt.title('Sinus-bølgen (ligning 1.8)')  # Add title to figure
plt.xlabel('x-verdier', fontsize=16)
plt.ylabel('y-verdier', fontsize=16)
plt.show()  # Show figure
#print(sine_wave.__doc__)
```

## Oppgave 1 - Lennard-Jones-potensialet

Lennard-Jones-potensialet blir ofte brukt som en approksimasjon for å illustrere interaksjonene mellom to nøytrale partikler, hvor kreftene mellom partiklene er frastøtende hvis partiklene er tett på hverandre. Potensialet illustrerer også at det finnes en likevekt-avstand mellom partiklene, hvor potensialet har sitt minimum. Lennard-jones potensialet kan uttrykkes som 

\begin{equation*}
V(r)=\epsilon((\dfrac{\sigma}{r})^{12} - 2(\dfrac{\sigma}{r})^{6}).
\end{equation*}

$r$ er avstanden mellom de to interagerende partiklene, $\epsilon$ er "styrken" på potensialet og $\sigma$ regulerer likevekt-avstanden mellom partiklene. I følgende oppgave skal du tenke på hva vi gjorde i introduksjonen når du lager scriptet ditt, og gi passende variabel- og funskjons-navn, samt kommenter koden din slik at andre kan forstå problemet du har løst. 

### Arbeidskrav:

- Lag en funksjon som regner ut Lennard-Jones potensialet. 
- Bruk en numpy array for å initialisere $r$-verdiene du bruker, f.eks. ved hjelp av numpy linspace.
- Regn ut potensialet for disse $r$-verdiene med passende verdier for $\epsilon$ og $\sigma$. 
- Finn $r$-verdien i arrayet med $r$-verdier som korresponderer til punktet hvor Lennard-Jones-potensialet har sitt minimum. Er det noen sammenheng mellom denne verdien for $r$ og $\epsilon, \sigma$?
- Bruk matplotlib til å illustrere potensialet, og indiker i plottet hvilken $r$-verdi som korresponderer til minimum av potensialet.

__Hint__: Potensialet er sensitivt for valg av parametere. Dette ser du hvis du f.eks. bruker $\sigma=10$ og har $r<1$ inkludert i arrayen med $r$-verdier. Hvis du får problemer med å illustrere potensialet i figuren din kan du prøve og bruke $\epsilon=6$, $\sigma=5$ og $r$-verdier mellom 4 og 9.

__Hint__: Det er mange måter å lagre figuren din på. Du kan gjøre det direkte fra matplotlib-figuren, eller du kan bruke `plt.savefig('min_figur.png')` for å lagre figuren direkte i scriptet ditt.


```python
from cProfile import label
import numpy as np
import matplotlib.pyplot as plt
"""SKRIV DIN LØSNING PÅ OPPGAVE 1 - Lennard-Jones-potensialet HER.
"""
def lennard_jones_potensial(r, epsilon, sigma):
    """ epsilon and sigma represents strength of the potential and an adjustment for equilibrium respectivliy. r is the distance"""
    return epsilon*((sigma/r)**12-2*(sigma/r)**6)

def main():
    r_values = np.linspace(4, 9)
    sigma = 5
    epsilon = 6

    potentials = lennard_jones_potensial(r_values, epsilon, sigma)
    min_potential, index = np.min(potentials), np.where(potentials == np.min(potentials))  # get minimum and corresponding index

    plt.plot(r_values, potentials, label="Lennard-jones-potential")
    plt.scatter(r_values[index], min_potential, label="minimum", color="r")
    plt.xlabel("distance")
    plt.ylabel("potential")
    plt.legend()
    # plt.show()
    plt.savefig("lennard_jones_potential.png")

main()
```

## Oppgave 2 - kule i fritt fall

For å finne farten som funksjon av tid for ei fallende kule, tar vi først utgangspunkt i N2,

\begin{equation*}
F=ma=m\dfrac{\text{d}v}{\text{d}t},
\end{equation*}

hvor $F$ er krafta, $m$ er massen og $a$ er akselerasjonen. Gitt vi har ei kule i fritt fall er krafta på kula gitt ved 

\begin{equation*}
F=mg,
\end{equation*}

hvor $m$ er kulas masse og $g$ er tyngdeakselerasjonen. Fra de to ovennevnte ligningene får vi 

\begin{equation*}
a = g = \dfrac{\text{d}v}{\text{d}t}
\end{equation*}

Med $a=g=-9.81$ m/$s^{2}$ er løsninga av differensialligninga for farten triviel, 

\begin{equation*}
v = \int_{0}^{t} g\text{d}t' = gt + v_{0},
\end{equation*}

hvor $v_{0}$ er initialfarten til kula. Men, i noen tilfeller (som du vil se i prosjektet som kommer senere), finnes det tilfeller hvor analytiske løsninger ikke er tilgjengelig. Da kan det være ønskelig å bruke en numerisk metode, som en approksimasjon, for å løse problemet. Metoden vi skal bruke for å beregne $v(t)$ er Eulers metode. Den baserer seg på en gitt tids-diskretisering, $\Delta t$. I det generelle tilfellet må $\Delta t$ være tilstrekkelig liten for at den numeriske løsninga skal bli presis. I vårt tilfelle, ved bruk av Eulers metode på differensialligninga, vil løsninga for farten se slik ut,

\begin{equation*}
v(t + \Delta t) \approx v(t) + \Delta t \dfrac{\text{d}v(t)}{\text{d}t}.
\end{equation*}

Hvis vi ønsker å evaluere denne ligningen for $t=10$ med $\Delta t = 1$ får vi altså

\begin{equation*}
v(11) \approx v(10) + \dfrac{\text{d}v(10)}{\text{d}t}.
\end{equation*}

Men, siden vi vet at $\dfrac{\text{d}v(t)}{\text{d}t} = g = -9.81$ m/$s^{2}$ for alle $t$, får vi 

\begin{equation*}
v(t + \Delta t) \approx v(t) - 9.81.
\end{equation*}

Hvis vi ønsker å bruke en numpy array for å beregne farten, kan Eulers metode se slik ut, hvor $i$ er indeksen i arrayet,

$v[i] = v[i-1] - 9.81 \Delta t$,

som holder for alle $i$, bortsett fra $i=0$. Med gitt $\Delta t = 1$ vil sammenhengen mellom indeksene og tiden $t$ bli $v[0] = v(t=0)$, $v[1] = v(t=1)$, $v[2] = v(t=2)$ osv.

### Arbeidskrav:

- Bruk ei for-løkke for å fylle `vel_arr` i kode-blokka under ved hjelp av Euler-metoden slik presentert ovenfor. Selv om eksempelkoden har $\Delta t = 1$ bør koden din kunne håndtere andre verdier for $\Delta t$.
- Bruk matplotlib og plot `vel_arr` som en funksjon av `times`.
- Gi figuren tittel og navn på aksene.
- `time_stop` er gitt til å være 10 sekunder. Regn ut den analytiske løsninga med ligninga gitt for farten ovenfor, $v = v_{0} + gt$, for $t = 10$ og sammenlign med resultatet du har fått med Euler-metoden. Hvis du bruker parameterene slik de er satt i kodeblokka under vil det siste elementet i `vel_arr` korrespondere til $t = 10$. 

```python
"""Code for obtaining the velocity as a function of time for a ball in free fall using Euler's method.
"""

import numpy as np
import matplotlib.pyplot as plt

g = -9.81  # Gravitational acceleration m/s^2 
init_vel = 10  # Initial velocity m/s

time_start = 3
time_stop = 10
delta_t = 2  # Time discretization
num_steps = int((time_stop - time_start) / delta_t) + 1# Number of steps from time_start to time_stop

times = np.linspace(time_start, time_stop, num_steps)  # array of times to evaluate from time_start to time_stop
vel_arr = np.zeros(len(times))  # Initialization of array for velocities
vel_arr[0] = init_vel  # v[0] = v(t = 0) = init_vel

"""LAG FOR-LØKKE FOR Å FYLLE fart_arr HER.
"""
for i in range(1, num_steps):
    vel_arr[i] = vel_arr[i-1] + delta_t*g   # euler method

"""PLOT RESULTATET MED MATPLOTLIB HER.
"""

plt.plot(times, vel_arr, label="fart")
plt.xlabel("tid")
plt.ylabel("fart")
plt.legend()
plt.show()

```

# Ekstraoppgave 1 - Modelering av skrått kast med luftmotstand

I denne oppgaven skal vi se på en kule med masse $m$ som blir kastet med en hastighet $\vec{v_0} = (v_{0x}, v_{0y})$ fra en høyde $h$. Som i oppgave 2 skal vi bruke Eulers metode til å regne ut posisjon som funksjon av tid. 

Summen av krefter på kulen er gitt ved Newtons andre lov: 

$\sum \vec{F} = m\vec{g} - b\vec{v}$, eller skrevet på komponent-form 

$ma_x = -b v_x$ 

$ma_y = - mg - bv_y$

Som i oppgave 2 kan vi diskretisere tiden ved å dele den inn i små intervaler med lengde $\Delta t$. Vi kan så bruke Eulers metode til å regne ut hastighet og posisjon s.f.a. tid ved hjelp av formlene
\begin{equation}
    v_x(t + \Delta t) = v_x(t) + a_x(t)\Delta t \\
    v_y(t + \Delta t) = v_y(t) + a_y(t)\Delta t
\end{equation}

\begin{equation}
    x(t + \Delta t) = x(t) + v_x(t)\Delta t \\
    y(t + \Delta t) = y(t) + v_y(t)\Delta t
\end{equation}

Oppgaven er å skrive et program som tar inn en utgangsposisjon $(x_0, y_0)$ og en utgangshastighet $(v_{0x}, v_{0y})$ og regner ut posisjon og hastighet som funksjon av tid frem til ballen treffer bakken $y=0$. Den skal også returnere tiden dette tar. Plott til slutt resultatene, dere kan sammenligne med den analytiske løsningen. Dere står fritt til å velge $m$ og $b$.

Hva skjer dersom dere øker utgangsposisjonen, blir hastigheten noen gang konstant?

Tips: For en tid $t$ skal vi regne ut hastighet, posisjon og akselerasjon for tiden $t+\Delta t$. Som vi ser fra ligningene er akselerasjonen ved tid $t$ avhengig av hastigheten, så det er viktig å regne ut disse i riktig rekkefølge. 

Tips: Det går ann å løse problemet analytisk for å dobbeltsjekke den numeriske løsningen.

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(
        delta_t: float,
        vel_0: tuple[float],
        pos_0: tuple[float],
        duration: float,
        m: float = 1,
        b: float = 0.5,
        g: float = 9.81
    ) -> tuple[tuple[float]]:
    """ 
    docs
    """
    vx_arr = np.zeros(duration)
    vy_arr =  np.zeros(duration)
    x_arr = np.zeros(duration)
    y_arr = np.zeros(duration)
    vx_arr[0], vy_arr[0] = vel_0
    x_arr[0], y_arr[0] = pos_0

    for i in range(1, duration):
        vx_arr[i] = vx_arr[i-1] - b / m * vx_arr[i-1] * delta_t
        vy_arr[i] = vy_arr[i-1] - (g + b / m * vy_arr[i-1]) * delta_t
        x_arr[i] = x_arr[i-1] + vx_arr[i-1]*delta_t
        y_arr[i] = y_arr[i-1] + vy_arr[i-1]*delta_t
    
    return (vx_arr, vy_arr), (x_arr, y_arr)

def main():
    delta_t = 0.1
    vel_0 = (12, 26)
    pos_0 = (0, 1)
    time_start = 0
    time_stop = 10
    duration = int((time_stop - time_start) / delta_t) + 1
    times = np.linspace(time_start, time_stop, duration)

    vel_arr, pos_arr = euler_method(delta_t, vel_0, pos_0, duration)

    plt.plot(times, vel_arr[0], label="x-fart")
    plt.plot(times, vel_arr[1], label="y-fart") 
    plt.xlabel("time")
    plt.ylabel("fart")
    plt.legend()
    plt.show()

    plt.plot(times, pos_arr[0], label="x-posisjon")
    plt.plot(times, pos_arr[1], label="y-posisjon")
    plt.xlabel("time")
    plt.ylabel("posisjon")
    plt.legend()
    plt.show()


main()
```
